# K-Nearest Neighbors

For this project, we are going to use the data set from sklearn "Breast Cancer"

```python
import seaborn
from sklearn.datasets import load_breast_cancer
```

### Data Exploration

```python
cancer_data = load_breast_cancer()
cancer_data.target_names
```

### Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(cancer_data.data, cancer_data.target, test_size=0.2, random_state=100)
```

### Run the KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
```

We are going to use a for loop to find the best k value

```python
k_list = []
for k in range(1, 101):
    classifier = KNeighborsClassifier(n_neighbors= k)
    classifier.fit(train_data, train_labels)
    accuracy = classifier.score(test_data, test_labels)
    k_list.append(accuracy)
```

### Graph the accuracy

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1, 101), k_list, color="red")
plt.title("Breast Cancer Classifier Accuracy")
plt.xlabel("K Value")
plt.ylabel("Validation Accuracy")
```
